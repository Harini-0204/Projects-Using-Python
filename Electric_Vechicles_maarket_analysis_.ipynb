 1. EV ADOPTION OVER THE TIME PERIOD

```python
sns.set_style("dark")
plt.figure(figsize=(14,7))
EV_Adoption_by_year = EV_Data['Model Year'].value_counts().sort_index()
sns.barplot(x=EV_Adoption_by_year.index, y=EV_Adoption_by_year.values,palette="dark")
plt.title("EV Adoption over the time period")
plt.xlabel('Model year')
plt.ylabel('Number of vehicles registered')
plt.show()
```

2. GEOGRAPHICAL DISTRIBUTION OF EV VEHICLES BY REGISTRATION

```python
county_ev_distribution=EV_Data['County'].value_counts()
top_counties=county_ev_distribution.head(3).index
top_three_counties=EV_Data[EV_Data['County'].isin (top_counties)]
ev_distribution_within_city=top_three_counties.groupby (['County','City']).size().sort_values(ascending=False).reset_index(name="No.of Vehicles")
top_city_distribution =ev_distribution_within_city.head(10)
plt.figure(figsize=(12, 8))
sns.set_style("dark")
sns.barplot(x='No.of Vehicles', y='City', hue= 'County',data=top_city_distribution, palette="icefire")
plt.title('Top 10 Geographical Distribution of EV Vehicles by registration')
plt.xlabel('No.of Vehicles Registered')
plt.ylabel('City')
plt.legend(title='County')
plt.show()
```

3. ANALYZING THE DISTRIBUTION OF EV VEHICLES BY ITS TYPES

```python
types_of_ev_distribution=EV_Data['Electric Vehicle Type'].value_counts()
types_of_ev_distribution.plot.pie(figsize=(12,9),autopct='%.1f')
plt.title('Distribution of EV by its types')
plt.legend(title='County')
plt.show()
```

4. DISTRIBUTION OF ELECTRIC RANGE

```python
plt.figure(figsize=(12,6))
sns.histplot(EV_Data['Electric Range'],bins=50,kde=True,color='mediumspringgreen')
plt.title('Distribution of EV range')
plt.xlabel('Electric range')
plt.ylabel('No.of.vechicles')
plt.axvline(EV_Data['Electric Range'].mean(),color='red', linestyle='--',label=f'Mean Range :{EV_Data["Electric Range"].mean():.2f} miles')
plt.legend()
plt.show()
```

5. AVERAGE ELECTIC RANGE BY MODEL YEAR


```python
Avg_range= EV_Data.groupby('Model Year')['Electric Range'].mean().reset_index()
plt.figure(figsize=(10,8))
sns.lineplot(x='Model Year', y= 'Electric Range', data=Avg_range,marker='o',color='maroon')
plt.title('Average EV Range by year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.show()
```
