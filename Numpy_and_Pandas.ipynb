```python
import numpy as np
import pandas as pd
```

```python
my_data=pd.DataFrame(np.arange(0,20).reshape(5,4),index=['a','b','c','d','e'],columns=[1,2,3,4]) #why it is not accepting rows as name in the place of index and also another name in the place of columns
print(my_data)
```

Indexing and slicing


```python
import numpy as np
import pandas as pd

my_data=pd.DataFrame(np.arange(0,20).reshape(5,4),index=['a','b','c','d','e'],columns=['col1','col2','col3','col4']) #why it is not accepting rows as name in the place of index and also another name in the place of columns
print(my_data)
```

```python
my_data.iloc[2:4, 2:]
```

```python
my_data.iloc[:,:]
```
